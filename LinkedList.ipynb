# Linked List in Python

A **Linked List** is a linear data structure, like arrays, that stores a collection of elements. However, unlike arrays, elements in a linked list are not stored at contiguous memory locations. Instead, they are linked using pointers.

### Key Concepts:
1.  **Node**: Each element in a linked list is called a node. A node typically consists of two parts:
    *   **Data**: The actual value stored in the node.
    *   **Next (or Pointer)**: A reference (or pointer) to the next node in the sequence. The last node's pointer is `None`.
2.  **Head**: The first node of the linked list.
3.  **Tail**: The last node of the linked list.

## 1. Node Class

The `Node` class represents an individual element in the linked list. Each node stores some `data` and a `next` pointer that points to the next node in the list. By default, the `next` pointer is initialized to `None`.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Example of creating a Node
node1 = Node(10)
print(f"Node data: {node1.data}, Next pointer: {node1.next}")
```

## 2. LinkedList Class

The `LinkedList` class manages the collection of nodes. It typically holds a reference to the `head` node of the list. We'll start with basic functionalities like initializing an empty list, adding elements to the end (append), and displaying the list.

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last_node = self.head
        while last_node.next:
            last_node = last_node.next
        last_node.next = new_node

    def display(self):
        current = self.head
        elements = []
        while current:
            elements.append(current.data)
            current = current.next
        print(" -> ".join(map(str, elements)))
```

### 2.1. Initialization and Appending Elements

Let's create a linked list and append some elements to it, then display its contents.

```python
my_list = LinkedList()
my_list.append(1)
my_list.append(2)
my_list.append(3)

print("Linked List after appending elements:")
my_list.display()

my_list.append(4)
print("Linked List after appending another element:")
my_list.display()
```

## 3. Extended LinkedList Class with More Operations

Let's extend the `LinkedList` class to include more common operations such as prepending, inserting at a specific position, deleting nodes, searching, getting the length, and reversing the list.

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last_node = self.head
        while last_node.next:
            last_node = last_node.next
        last_node.next = new_node

    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_at_position(self, data, position):
        if position < 0:
            print("Invalid position. Position cannot be negative.")
            return
        if position == 0:
            self.prepend(data)
            return

        new_node = Node(data)
        current = self.head
        count = 0
        while current and count < position - 1:
            current = current.next
            count += 1

        if current is None:
            print("Position out of bounds. Appending to end.")
            self.append(data)
        else:
            new_node.next = current.next
            current.next = new_node

    def delete_by_value(self, key):
        current = self.head

        # Case 1: Head node itself holds the key
        if current and current.data == key:
            self.head = current.next
            current = None
            return

        # Case 2: Search for the key, keeping track of the previous node
        prev = None
        while current and current.data != key:
            prev = current
            current = current.next

        # If key was not present in linked list
        if current is None:
            print(f"Key {key} not found in the list.")
            return

        # Unlink the node from linked list
        prev.next = current.next
        current = None

    def delete_by_position(self, position):
        if self.head is None:
            print("List is empty, nothing to delete.")
            return

        # Case 1: Delete the head node
        if position == 0:
            self.head = self.head.next
            return

        # Search for the node to be deleted, keeping track of the previous node
        current = self.head
        count = 0
        prev = None

        while current and count < position:
            prev = current
            current = current.next
            count += 1

        # If position is greater than number of nodes
        if current is None:
            print(f"Position {position} out of bounds.")
            return

        # Unlink the node from linked list
        prev.next = current.next

    def search(self, key):
        current = self.head
        while current:
            if current.data == key:
                return True
            current = current.next
        return False

    def get_length(self):
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        return count

    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    def display(self):
        current = self.head
        elements = []
        while current:
            elements.append(current.data)
            current = current.next
        print(" -> ".join(map(str, elements)))
```

### 3.1. Prepend (Add to the beginning)

```python
my_list = LinkedList()
my_list.append(2)
my_list.append(3)
print("Original list:")
my_list.display()

my_list.prepend(1)
print("List after prepending 1:")
my_list.display()
```

### 3.2. Insert at a specific position

```python
my_list = LinkedList()
my_list.append(1)
my_list.append(3)
my_list.append(4)
print("Original list:")
my_list.display()

my_list.insert_at_position(2, 1) # Insert 2 at index 1
print("List after inserting 2 at position 1:")
my_list.display()

my_list.insert_at_position(5, 4) # Insert 5 at end
print("List after inserting 5 at position 4 (end):")
my_list.display()
```

### 3.3. Delete by Value

```python
my_list = LinkedList()
my_list.append(1)
my_list.append(2)
my_list.append(3)
my_list.append(4)
print("Original list:")
my_list.display()

my_list.delete_by_value(3)
print("List after deleting 3:")
my_list.display()

my_list.delete_by_value(1)
print("List after deleting 1 (head):")
my_list.display()

my_list.delete_by_value(5) # Trying to delete a non-existent value
print("List after trying to delete 5:")
my_list.display()
```

### 3.4. Delete by Position

```python
my_list = LinkedList()
my_list.append(10)
my_list.append(20)
my_list.append(30)
my_list.append(40)
print("Original list:")
my_list.display()

my_list.delete_by_position(0) # Delete head
print("List after deleting node at position 0:")
my_list.display()

my_list.delete_by_position(1) # Delete node at index 1 (which is 30 now)
print("List after deleting node at position 1:")
my_list.display()

my_list.delete_by_position(5) # Out of bounds position
print("List after trying to delete node at position 5:")
my_list.display()
```

### 3.5. Search for an element

```python
my_list = LinkedList()
my_list.append(100)
my_list.append(200)
my_list.append(300)
print("List:")
my_list.display()

print(f"Is 200 in the list? {my_list.search(200)}")
print(f"Is 500 in the list? {my_list.search(500)}")
```

### 3.6. Get Length of the List

```python
my_list = LinkedList()
my_list.append('A')
my_list.append('B')
my_list.append('C')
print("List:")
my_list.display()
print(f"Length of the list: {my_list.get_length()}")

empty_list = LinkedList()
print(f"Length of an empty list: {empty_list.get_length()}")
```

### 3.7. Reverse the List

```python
my_list = LinkedList()
my_list.append(1)
my_list.append(2)
my_list.append(3)
my_list.append(4)
print("Original list:")
my_list.display()

my_list.reverse()
print("List after reversing:")
my_list.display()

empty_list = LinkedList()
print("Empty list before reversing:")
empty_list.display()
empty_list.reverse()
print("Empty list after reversing:")
empty_list.display()
```
